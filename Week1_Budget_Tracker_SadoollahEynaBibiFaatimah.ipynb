## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### üí° Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
Use python to write this specific code. Allows users to input their monthly income. Allows users to input up to 3 expenses. Calculates remaining budget. Shows a warning if budget is negative.
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Ask user for monthly income
income = float(input("Enter your monthly income: "))

# Ask user to input up to 3 expenses
expense1 = float(input("Enter expense 1: "))
expense2 = float(input("Enter expense 2: "))
expense3 = float(input("Enter expense 3: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display remaining budget
print("\nTotal expenses:", total_expenses)
print("Remaining budget:", remaining_budget)

# Show warning if budget is negative
if remaining_budget < 0:
    print("Warning: Your budget is negative! You are overspending.")
else:
    print("Good job! You are within your budget.")
```

### üêõ Debugging and Refinement

```python
# Document any issues and fixes
issues_found = [
    "Issue 1: Program crashed when user entered text instead of numbers.",
    "Fix 1: Converted inputs to float() to ensure numerical calculations.",

    "Issue 2: Remaining budget calculation was incorrect due to missing one expense.",
    "Fix 2: Added all three expenses to calculate total_expenses correctly.",

    "Issue 3: No warning displayed when budget was negative.",
    "Fix 3: Added an if statement to check if remaining_budget < 0 and display warning."

]

print("üîß DEBUGGING NOTES:")
for item in issues_found:
    print(f"- {item}")
```

```python
# REFINED VERSION - Your Working Budget Tracker
print("=== Monthly Budget Tracker ===")

# Get monthly income with validation
while True:
    try:
        income = float(input("Enter your monthly income: "))
        if income < 0:
            print("Income cannot be negative. Please try again.")
        else:
            break
    except ValueError:
        print("Invalid input. Please enter a numeric value.")

# Get up to 3 expenses
expenses = []

for i in range(1, 4):
    while True:
        try:
            expense = float(input(f"Enter expense {i} (or 0 if none): "))
            if expense < 0:
                print("Expense cannot be negative. Please try again.")
            else:
                expenses.append(expense)
                break
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

# Calculate totals
total_expenses = sum(expenses)
remaining_budget = income - total_expenses

# Display results
print("\n=== Budget Summary ===")
print(f"Monthly Income: {income}")
print(f"Total Expenses: {total_expenses}")
print(f"Remaining Budget: {remaining_budget}")

# Warning if negative budget
if remaining_budget < 0:
    print("‚ö† WARNING: Your budget is negative! You are overspending.")
else:
    print("‚úÖ Good job! You are within your budget.")
```
