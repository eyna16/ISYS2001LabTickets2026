## Step 5 – Write the Code

Translate your pseudocode into Python code. Below is one example solution:

````python
# Import datetime to timestamp your work
from datetime import datetime

# Step 5 – Python Code: Personal Finance Tracker Profile Setup

# Start the program

# Prompt the user to enter their name and store it
user_name = input("Enter your name: ")

# Prompt the user to enter their preferred theme color and store it
theme_color = input("Enter your preferred theme color (e.g., blue, green, dark): ")

# Prompt the user to enter their primary currency and store it
primary_currency = input("Enter your primary currency (e.g., USD, AUD, EUR): ")

# Prompt the user to enter their main savings goal and store it
main_savings_goal = input("Enter your main savings goal (e.g., New Car, Holiday): ")

# Construct a welcome and summary message
summary = (
    f"Welcome to your Personal Finance Tracker, {user_name}! "
    f"Your profile is ready. The app will use a {theme_color} theme "
    f"and all amounts will be in {primary_currency}. "
    f"Let's start working towards your goal of saving for a {main_savings_goal}!"
)

# Print the summary message to the user
print(summary)

# End the program
````
