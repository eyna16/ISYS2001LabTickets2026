## Part 2: Financial Decision Logic (15 minutes)

### Budget Limit Checker
Apply your if/else skills to real budget management:

````python
print("=== Budget Limit Checker ===")

# Get user's budget and expense
monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter expense amount: $"))

#formula to calculate percentage of budget
percentage_of_budget = (expense/ monthly_budget) * 100

# Decision logic here - you write this part!
# If expense > 50% of budget: "Major purchase - think carefully"
# If expense > 25% of budget: "Significant expense"
# If expense > 10% of budget: "Moderate expense"
# Otherwise: "Minor expense - within budget"

if percentage_of_budget >=  50:
  category = "Major purchase - think carefully"
  percentageamount= "50%"
elif percentage_of_budget >= 25:
  category = "Significant expense"
  percentageamount= "25%"
elif  percentage_of_budget >= 10:
  category = "Moderate expense"
  percentageamount= "10%"
else:
  category = "Minor expense - within budget"
print(f"This is {category} You are at {percentageamount} ")



````

### Savings Goal Progress
Use multiple conditions to track savings progress:

````python
print("=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))

# Your task: Calculate percentage and give encouraging feedback
# 100%+: "Goal achieved! Consider a new goal."
# 75-99%: "Almost there! You're doing great."
# 50-74%: "Halfway there - keep it up!"
# 25-49%: "Good progress, stay consistent"
# Under 25%: "Just getting started - you've got this!"

percentageofgoal = (current_savings/goal_amount) *100

if percentageofgoal >= 100:
  goal = "Goal achieved! Consider a new goal."
elif percentageofgoal >= 75 and percentageofgoal <= 99:
  goal = "Almost there! You're doing great."
elif percentageofgoal >= 50 and percentageofgoal <= 74:
  goal = "Halfway there - keep it up!"
elif percentageofgoal >= 25 and percentageofgoal <= 49:
  goal = "Good progress, stay consistent"
else:
  goal = "Just getting started - you've got this!"
print(goal)

````

## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
print("=== Smart Purchase Advisor ===")

# Get financial information
available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Calculate affordability percentage
# How much of your available money this purchase would use (as a percentage)
affordability_ratio = purchase_price / available_money *100

#calculate income ratio
# This shows the user's monthly income the purchase would take
income_ratio = purchase_price / monthly_income * 100

# Your task: Build the decision logic
# Consider:
# - Can they afford it at all?
# - Is it a large percentage of their available money?
# - Is it reasonable compared to their income?
# - What advice should you give?

# Suggestion: Use multiple if/elif conditions to give personalised advice


if purchase_price > available_money:
  shortfall = purchase_price - available_money
  print(f"Save up first - you need ${shortfall:.2f} more.")

elif affordability_ratio > 70:
  print(f"You can afford it, but is is {affordability_ratio:.1f}% of your available money. Consider waiting.")

elif income_ratio > 50:
  print(f"This purchase if {income_ratio:.1f}% of your monthly income. Make sure it won't affect essentials.")

elif affordability_ratio <= 10:
  print(" This is a minor expense - go for it!")

else:
  print("This seems like a resonable purchase for your budget.")



#Emergency fund check
emergency_fund = float(input("How much money do you have in your emergency fund? $"))

if (available_money - purchase_price) < (0.3 * emergency_fund):
  print("Warning: This purchase may drain your emergency savings.")

else:
  print("Your emergency fund is safe after this purchase.")


#Category-based advice

#ask for category(luxury or necessity)
category = str(input("Is this item a necessity or a luxury?"))

if category == "necessity":
 print("Since this is a necessity, it is resonable to buy it if you can afford it.")

elif category == "luxury":
 if affordability_ratio > 50:
  print("This is a luxury item and takes a large portion of your money. Consider waiting.")

 else:
  print("Luxury item looks okay, but make sure it does not affect your budget.")

else:
  print("Unknown category. Proceed with caution!")

#Payment plan calculator

if purchase_price > 0.5 * monthly_income:
    months = int(input("Do you want to see a payment plan? Enter number of months: "))
    monthly_payment = purchase_price / months
    print(f"You could pay ${monthly_payment:2f} per month over {months} months.")
````

Create a new notebook called "Personal_Finance_Tracker" and build the foundation:

````python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# This week's feature: Expense categorisation
print("\n--- Expense Entry ---")
expense_amount = float(input("Enter an expense: $"))
expense_description = input("What was this expense for? ")

# Apply your decision-making skills here
# Categorise the expense and give feedback
# This becomes the foundation you'll build on each week!

if expense_description.lower() in ["rent", "utilities", "groceries", "transport"]:
  category = "Essentials"

elif expense_description.lower() in ["movie", "dining out", "games", "shopping"]:
  category = "Entertainment"

elif expense_description.lower() in ["investment", "savings", "retiremnet fund"]:
  category = "Savings"

else:
  category = "Others"

expense_percentage = (expense_amount /monthly_income) * 100

print(f"\nExpense Summary:")
print(f"- Description: {expense_description}")
print(f"- Amount: ${expense_amount:.2f}")
print(f"- Category: {category}")
print(f"- This expense is{expense_percentage:.2f}% of your monthly income.")

if expense_percentage > 50:
  print("Warning: This expense is a large portion of your income!")

elif expense_percentage > 20:
  print("Note: Consider if this expense is necessary or can be reduced.")

else:
  print("Good job! this expense is within a resonable range.")
````
